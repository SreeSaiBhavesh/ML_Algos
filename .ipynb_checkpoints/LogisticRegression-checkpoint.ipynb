```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('heart_disease_dataset_UCI.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data = data.dropna(axis=0)
```

```python
data.isnull().sum()
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_regressor = LogisticRegression()
logistic_regressor.fit(X_train, y_train)
```

```python
y_pred_lgr = logistic_regressor.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
print("Accuracy Score is: ", accuracy_score(y_test,y_pred_lgr))
```

```python
print("Error rate is: ", 1-(accuracy_score(y_test,y_pred_lgr)))
```

```python
print("Precision Score: ", precision_score(y_test,y_pred_lgr))
```

```python
print("Recall Score: ",recall_score(y_test, y_pred_lgr))
```

```python
from sklearn.metrics import confusion_matrix as cm
c = cm(y_test,y_pred_lgr)
```

```python
sensitivity1 = c[0,0]/(c[0,0]+c[0,1])
print('Sensitivity : ', sensitivity1 )

specificity1 = c[1,1]/(c[1,0]+c[1,1])
print('Specificity : ', specificity1)
```
